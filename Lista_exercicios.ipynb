```
#Exercício 1:
print('='*30)
print('Exercício 1:')
print('='*30)

import numpy as np
from numpy import matrix as matrix
from scipy import linalg

#Matriz (A) a ser trabalhada:
A = [[7,2,1],[0,3,-1],[-3,4,-2]]
print('-'*30)

#Printa a matriz:
print('A matriz é:')
for l in range(0,3):
  for c in range(0,3):
    print(f'[{A[l][c]}]', end='')
  print()
print('-'*30)

#Calcula a inversa:
print('A inversa de A é:')
inversa = linalg.inv(A)
for l2 in range(0,3):
  for c2 in range(0,3):
    print(f'[{inversa[l2][c2]}]', end='')
  print()
print('-'*30)

#Calcula o determinante de A:
print('O determinante de A é:')
p = linalg.det(A)
print(p)
print('-'*30)

#Calcula o traço da matriz A:
print('O traço de A é:')
traço = np.trace(A)
print(traço)
print('-'*30)

#Exercício 2:
print('='*30)
print('Exercício 2:')
print('='*30)
print('-'*30)

#Define matriz b para o exercício 2:
b = matrix([[21],[5],[-1]])
print('A matriz b é:')
print(b)
print('-'*30)

#Solução de Ax = b:
#print('A solução de Ax = b, por multiplicação matricial é:')
print('Por produto matricial:')
P2 = matrix([[7,2,1],[0,3,-1],[-3,4,-2]])
print(P2.I*b)

print('Utilizando a função do numpy, encontramos:')
s = np.linalg.solve(A,b)
print(s)
print('-'*30)
```
