```
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from math import *

data = 'dados2.txt'
data2 = 'dados.txt'

t2 = np.loadtxt(data2, usecols=0)
E2 = np.loadtxt(data2, usecols=1)
EV_2 = 80.2

t = np.loadtxt(data, usecols=0)
E = np.loadtxt(data, usecols=1)

L = 156
EV_0 = 106

def func(L,E,EV_0):
  return (1/L)*np.log(E/EV_0)

def func2(L,E2, EV_2):
  return (1/L)*np.log(E2/EV_2)  

plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(10,6))
x = np.linspace(-0.50, 1., 1000)

# Outer Plot
ax.plot(t2, func2(E2,L,EV_2),'s', label = 'Coeficiente de atenuação - laser $\sigma = 0,00489 cm^{-1}$', color = 'red')
ax.set(xlim=(0, 40), ylim=(0,0.9))
fig.tight_layout()
plt.errorbar(t2,func2(E2,L,EV_2),yerr=0.004897, color='red')
plt.xlabel('Tempo (min)')
plt.ylabel('Coeficiente de atenuação $(cm^{-1})$')
plt.legend()

# Inner Plot
inner_ax = fig.add_axes([0.2, 0.55, 0.35, 0.35]) # x, y, width, height
inner_ax.plot(t, func(E,L,EV_0),'s', label = 'Coeficiente de atenuação - lâmpada $\sigma = 0,004690 cm^{-1}$', color = 'green')
inner_ax.set(title='', xlim=(0, 8), ylim=(0, .013))
ax.set_title("Coeficiente de atenuação em função do tempo", fontsize=20)
plt.errorbar(t, func(E,L,EV_0), yerr=0.004690, color = 'green')
plt.xlabel('Tempo (min)')
plt.ylabel('Coeficiente de atenuação $(cm^{-1})$')
plt.legend()
plt.show()
```
