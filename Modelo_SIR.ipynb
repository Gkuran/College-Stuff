```
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt

ndays = 100
dt = 1
beta = (1/3)
gamma = (1/14)

S = np.zeros(ndays)
I = np.zeros(ndays)
R = np.zeros(ndays)
t = np.arange(ndays)*dt

I[0] = 0.1
S[0] = 1 - I[0]
R[0] = 0

N = S + I + R


for i in range(ndays - 1):
  S[i+1] = S[i] - beta*(S[i]*I[i])*dt
  I[i+1] = I[i] + (beta*S[i]*I[i] - gamma*I[i])*dt
  R[i+1] = R[i] + (gamma*I[i])*dt

plt.plot(t, S, 'r', lw =1, label = 'Susceptíveis')
plt.plot(t, I, 'g', lw =1, label = 'Infectados')
plt.plot(t, R, 'b', lw =1, label = 'Removidos')
plt.xlabel('Tempo (dias)')
plt.ylabel('População (Normalizada)')
plt.legend()

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('classic')

gamma = 0.5


beta  = 15
S     = 80.
I     = 5.
I2 =  5.
R     = 15.
tf    = 25
t = 0

dt = 0.01

N = S + I + R


N_list = [N]
S_list = [S]
I_list = [I]
R_list = [R]
t_list = [t]

while t < tf:
  S = S - (beta*I*S*dt)/N
  I = I + (beta*I*S*dt)/N - gamma*I*dt
  R = R + gamma*I*dt
  N = S + I + R

  t += dt  


  N_list.append(N)
  S_list.append(S)
  I_list.append(I)
  R_list.append(R)
  t_list.append(t)

plt.plot(t_list, S_list, 'blue', label = 'Susceptíveis$\;(S = 80)$')
plt.plot(t_list, I_list, 'red', label = 'Infectados$\;(I = 5)$')
plt.plot(t_list, R_list, 'green', label = 'Removidos$\;(R = 15$)')
plt.title('Modelo SIR: Caso III')
plt.xlabel('Tempo (dias)')
plt.ylabel('População')
plt.legend()
plt.show()
```
