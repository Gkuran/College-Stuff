```
#PLOT DOS GRÁFICOS x X t

import matplotlib.pyplot as plt
import numpy as np

x  =   float(input('valor de x inicial: '))
v  =   float(input('valor de v inicial: '))
A  =   float(input('Valor da amplitude:'))
w  =   1
tf =   24
dt =   0.1
t  =   0

t_list  = [0]
x_list  = [x]
v_list  = [v]
xa_list = [x]
va_list = [v]


while t <= tf :
  t += dt
  x = x + v*dt
  v = v - w**2*np.sin(x)*dt
  t_list.append(t)
  x_list.append(x)
  v_list.append(v)
  xa_list.append(A*np.cos(w*t))
  va_list.append(-A*w*np.sin(w*t))

plt.style.use('dark_background')

plt.plot(t_list, x_list, label=A)
plt.plot(t_list, xa_list, label='Solução Analítica')
plt.legend(loc='upper right')
plt.ylabel(r'$\theta$')
plt.xlabel('Tempo')
plt.show()
```

```
# ESPAÇO DE FASE

import matplotlib.pyplot as plt
import numpy as np

tf = 24
dt = 0.1
w = 1

for i in range(6):

  x0 = [0.1,0.5,1,3,9.5,-9.5]
  x = [0.1,0.5,1,3,9.5,-9.5]
  v = [0,0,0,0,-0.1,0.1]

  t = 0
  t_list=[0]
  x_list=[x[i]]
  v_list=[v[i]]

  while t < tf:

    t += dt

    x[i] = x[i] + v[i]*dt
    v[i] = v[i] - w*np.sin(x[i])*dt

    t_list.append(t)
    x_list.append(x[i])
    v_list.append(v[i])

  plt.style.use('dark_background')

  plt.plot(x_list,v_list, label = str(x0[i]))
  plt.title('Espaço de Fase')
  plt.xlabel(r'$\theta$')
  plt.ylabel(r'$\dot{\theta}$')
  plt.legend()
  
```

```
# ENERGIAS

import matplotlib.pyplot as plt
import numpy as np


w  =   1
tf =   24
dt =   0.1
t  =   0
l = 10
g = 10
m = 1


x0 = [0.1,0.5,1,3,9.5,-9.5]
x = [0.1,0.5,1,3,9.5,-9.5]
v = [0,0,0,0,-0.1,0.1]


for i in range(6):
  

  t=0
  E=0
  t_list = [0]
  x_list = [x[i]]
  v_list =[v[i]]
  E_list = [m*(v[i]**2*l**2)/2+m*g*l*(1-1*np.cos(x[i]))]

  while t < tf:

    t += dt
    x[i] = x[i] +v[i]*dt
    v[i] = v[i] - w**2*np.sin(x[i])*dt
    
    E = m*(v[i]**2*l**2)/2+m*g*l*(1-1*np.cos(x[i]))

    t_list.append(t)
    x_list.append(x[i])
    v_list.append(v[i])
    E_list.append(E)

  plt.style.use('dark_background')

  plt.plot(t_list,E_list, label = str(x0[i]))
  plt.xlabel('Tempo')
  plt.ylabel('Energia')
  plt.title('Energia Total')
  plt.legend()

  
```
